```python
import requests

API_KEY = "7865948ab6f0acd5a5ae7284709f3ffe"  

def get_weather_data(city_name):
    url = f"http://api.openweathermap.org/data/2.5/weather?q={city_name}&appid={API_KEY}&units=metric"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        return data
    else:
        return None

def get_weather_for_date(city_name, date):
    weather_data = get_weather_data(city_name)

    if weather_data:
        print(f"Weather Report for {city_name} on {date}:")
        print(f"Temperature: {weather_data['main']['temp']}Â°C")
    else:
        print("Error fetching weather data. Please check your API key or city name.")

def get_wind_speed_for_date(city_name, date):
    weather_data = get_weather_data(city_name)

    if weather_data:
        print(f"Wind Speed for {city_name} on {date}: {weather_data['wind']['speed']} m/s")
    else:
        print("Error fetching weather data. Please check your API key or city name.")

def get_pressure_for_date(city_name, date):
    weather_data = get_weather_data(city_name)

    if weather_data:
        print(f"Pressure for {city_name} on {date}: {weather_data['main']['pressure']} hPa")
    else:
        print("Error fetching weather data. Please check your API key or city name.")

def main():
    while True:
        print("\nSelect an option:")
        print("1. Get weather")
        print("2. Get Wind Speed")
        print("3. Get Pressure")
        print("0. Exit")

        choice = input("Enter your choice: ")

        if choice == '1':
            city_name = input("Enter the city name: ")
            date = input("Enter the date (YYYY-MM-DD): ")
            get_weather_for_date(city_name, date)
        elif choice == '2':
            city_name = input("Enter the city name: ")
            date = input("Enter the date (YYYY-MM-DD): ")
            get_wind_speed_for_date(city_name, date)
        elif choice == '3':
            city_name = input("Enter the city name: ")
            date = input("Enter the date (YYYY-MM-DD): ")
            get_pressure_for_date(city_name, date)
        elif choice == '0':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```
